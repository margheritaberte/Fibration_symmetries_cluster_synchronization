### Computation of fibre partition for real dataset with multi-body interactions.

The datasets are taken from https://github.com/xgi-org/xgi-data.

```python
# Libraries
import sys
sys.path.insert(0, '../Utilities')

from networkx.algorithms import bipartite
import xgi

from utilities_fibration import *
```

```python
hg_dataset_lists = [
    'diseasome',
    ####'disgenenet',
    ###'hospital-lyon',
    ###'house-committees',
    ###'hypertext-conference',
    ###'invs13',
    ###'malawi-village',
    ###'ndc-classes',
    ###'ndc-substances',
    ###'plant-pollinator-mpl-015',
    ###'plant-pollinator-mpl-016',
    ###'plant-pollinator-mpl-049',
    ###'plant-pollinator-mpl-062',
    ###'science-gallery',
    ###'senate-committees',
]
```

```python
for hg in hg_dataset_lists:
    print(hg)

    H = xgi.load_xgi_data(hg)
    num_nodes = H.num_nodes
    num_edges = H.num_edges

        
    if (num_nodes + num_edges < 165000): # Avoid too large hypergraphs

        H.cleanup(singletons=False, isolates=False, connected=True, relabel=False)
        
        num_nodes = H.num_nodes
        num_edges = H.num_edges

        print(f'num_nodes: {num_nodes}')
        print(f'num_edges: {num_edges}')

        G = xgi.to_bipartite_graph(H)
        print('Bipartite done!')

        bsc = G[0]
        color_dict_iter = bipartite.color(G)
        #print(color_dict_iter)
        #color_dict_list.append(bhg_dict)

        result_sc = kameicock_coloring(G, color_dict_iter)
        #print("Final coloring:", result_sc) 

        nodes_layer1 = [n for n, l in color_dict_iter.items() if l == 1]
        nodes_coloring = {n: c for n, c in result_sc.items() if n in nodes_layer1}

        cl_hg = []
        for val in set(nodes_coloring.values()):
            cl_hg.append([k for k, v in nodes_coloring.items() if v == val])

        print(f'n nodi layer 1: {len(nodes_layer1)}')
        print(f'ratio nodes/clusters: {len(nodes_layer1)/len(cl_hg)}')

        

        c = [e for e in cl_hg if len(e) > 1]
        print(f'Number of non trivial clusters: {len(c)}')
        print('**************************\n')
    

    else:
        print(f'Too many/little edges and nodes: {num_nodes + num_edges}')
        print('**************************\n')
    
```
